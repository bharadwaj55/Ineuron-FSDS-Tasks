```python
# Drop the rows which are not having atleast 3 'non-Nan' values in a row
# It checks for minimum 'Threshold' of 3
# If there are 'non-Nan' values < less than 'threshold' 3, then deletes that row
df3.dropna(thresh=3)
```

```python
# Fill Nan values with any value
df3.fillna(value=4)
```
